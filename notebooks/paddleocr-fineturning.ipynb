```python
import pandas as pd
import os
import sys
sys.path.append('/kaggle/working/compare_ocr_benchmark/common')
from metrics import compute_metrics
from utils import save_results
```

```python
# Đường dẫn dữ liệu và annotation
CSV_ANN = '/kaggle/input/nckh-2425-crops/crops_gt.csv'
PRED_TXT_PATH = '/kaggle/working/PaddleOCR/output/rec/predicts_ppocrv3_latin.txt'  
OUT_CSV = '/kaggle/working/compare_ocr_benchmark/results/paddleocr_results.csv'
```

```python
# 1. Đọc file predict PaddleOCR
results = []
with open(PRED_TXT_PATH, encoding='utf-8') as f:
    for line in f:
        if '\t' in line:
            fname, pred = line.strip().split('\t', 1)
            # Nếu pred có tab (text + confidence), chỉ lấy phần text
            pred = pred.split('\t')[0].strip()
            results.append({'filename': os.path.basename(fname), 'predicted_text': pred})
df_pred = pd.DataFrame(results)


# 2. Đọc ground truth
df_gt = pd.read_csv(CSV_ANN)

# 3. Merge theo filename
df_merged = pd.merge(df_gt, df_pred, on='filename', how='left')

# 4. Tính metrics với hàm chung đã có
all_results = []
for idx, row in df_merged.iterrows():
    desc_gt = str(row['description_gt'])
    pred = str(row['predicted_text']) if not pd.isna(row['predicted_text']) else ""
    label = row['label'] if 'label' in row else ""

    metrics = compute_metrics(desc_gt, pred)
    all_results.append({
        "filename": row['filename'],
        "label": label,
        "ground_truth": desc_gt,
        "predicted_text": pred,
        "cer": metrics["cer"],
        "wer": metrics["wer"],
        "lev": metrics["lev"],
        "acc": metrics["acc"]
    })
    if idx % 50 == 0:
        print(f"Processed {idx}/{len(df_merged)}")
```

```python
os.makedirs(os.path.dirname(OUT_CSV), exist_ok=True)
save_results(all_results, OUT_CSV)
```
